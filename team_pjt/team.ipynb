```python
#전체 사례수 성별 분포
import pandas as pd
import numpy as np
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False #한글 사용
Book_Read = pd.read_excel("#210075_2021년 국민독서실태조사_성인_Table.xlsx",index_col=0)
#print(Book_Read)
#print("="*50)
Total = Book_Read.iloc[:,:3]
Student = Total.iloc[9:12]
Income = Total.iloc[19:24]
Total.head(4)
```

```python
#학력별 사례수 퍼센트 데이터 시각
import matplotlib.pyplot as plt
import seaborn as sns
ratio = [900, 2136,  2964]
labels = ["중졸이하", "고졸/고퇴", "대재이상"]
wedgeprops = {"width":0.8, "edgecolor": "k", "linewidth" : 1}
plt.pie(ratio, labels = labels, shadow=True, autopct="%.1f%%", wedgeprops=wedgeprops)
plt.show()
print("="*50)
```

```python
#평일, 휴일 학력별 (종이,전자,오디오)책 독서 시간(분)
A = Book_Read.iloc[787:790]
Student_Read_Time = A.iloc[:,:8]
Student_Read_Time.head()
#Unnamed1 학력, Unnamed2 사례수, Unnamed3 종이-평일(분), Unnamed4 전자=평일(분), Unnamed5 오디오=평일(분)
#Unnamed6 종이=휴일(분), Unnamed7 전자=휴일(분), Unnamed8 오디오=휴일(분) 
```

```python
#학력별 총 독서시간(분) 데이터 시각
Read_Mid = 4.0+0.1+0.0+2.5+0.1+0.0
Read_High = 9.8+1.9+0.5+10.6+2.5+0.5
Read_Uni = 16.9+12.0+2.5+23.7+18.0+2.8
x = np.arange(3)
plt.bar(x,[Read_Mid,Read_High,Read_Uni], color=["r","g","b"],  width=0.6)
plt.xticks(x,["Read_Mid", "Read_High", "Read_Uni"])
plt.grid(True, axis="y")
plt.title("Student_Read_Time",loc="center",pad=20)
plt.show()
#학력과 독서량은 비례한다는 가정 True
```
