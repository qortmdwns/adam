```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(3)
plt.bar(x,[100,400,900], color=["r","g","b"],width=0.6)
plt.xticks(x,["2018", "2019", "2020"])
plt.show()
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)
plt.scatter(x,y,s=area, c=colors, cmap="Spectral",alpha=0.5)
plt.show()
```

```python
ratio = [34,32,16,18]
labels = ["Apple", "Banana", "Melon", "Grapes"]
wedgeprops = {"width":0.7, "edgecolor":"k", "linewidth": 1}
plt.pie(ratio,labels=labels,autopct="%.1f%%", explode=[0,0.1,0,0.1], colors=["red","yellow","green","purple"],wedgeprops=wedgeprops)
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov" , "Dec"]
sales_2019 = [100, 120, 140, 110, 130, 150, 160, 170, 180, 200, 190, 210]
sales_2020 = [90, 110, 130, 120, 140, 160, 170, 160, 150, 180, 200, 190]
plt.plot(months, sales_2019, label = "sales_2019")
plt.plot(months, sales_2020, label = "sales_2020")
plt.title("Monthly Sales Comparison(2019-2020)",loc="center")
plt.xlabel("months", loc="center")
plt.ylabel("sales", loc="center")
plt.legend(loc="best",ncol=1,fontsize=10,frameon=True,shadow=True)
plt.show()


```

```python
categories = ["Categories1", "Categories2", "Categories3", "Categories4", "Categories5"]
data = [20, 35, 15, 27, 45]
x = np.arange(5)
plt.bar(x,[20, 35, 15, 27, 45], width=0.6)
plt.xticks(x,["Categories1", "Categories2", "Categories3", "Categories4", "Categories5"])
plt.title("Bar Chart",loc="center")
plt.xlabel("Categories", loc="center")
plt.ylabel("Values", loc="center")
plt.grid(True, axis="both")
plt.show()
```

```python
labels = ["Apple", "Banana", "Mango", "Blueberry"]
sizes = [15,30,45,10]
wedgeprops = {"width":1, "linewidth": 1}
explode = (0, 0.1,0, 0)
plt.pie(sizes, labels=labels,autopct="%.1f%%", startangle=90, shadow=True, explode=[0, 0.1, 0, 0] ,wedgeprops=wedgeprops)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_axis = np.linspace(0,10,200) # 0~10 200
sin_1 = np.sin(x_axis)
cos_1 = np.cos(x_axis)
plt.title('Sine and Cosine Functions', loc='center', pad=20)
plt.plot(x_axis,sin_1, label = 'sin(x)',color = 'blue')
plt.plot(x_axis,cos_1, label = 'cos(x)',color = 'red')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, axis = 'both')
plt.legend(loc = 'best', ncol =1, fontsize = 10, frameon = True, shadow = True)
plt.show()
```
